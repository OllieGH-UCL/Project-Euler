**Colour Space**

- A new color can be produced by mixing two paints if and only if it, in the color-space, lies on the segment defined by the other 2 paints.
- A new color can be produced by mixing 3 paints if and only if it, in the color-space, lies in the triangle defined by their other 3 paints.
- The Mixture ratio is inversely determined by a geometric measure. Co-linear will be length of segment. Tri will be the area of the triangle formed.

_Question_ - We now have three kinds of paint, which are U = (10%, 10%), V = (60%, 10%), W = (10%, 30%). Which of the following colors can't be generated by mixing these 3 kings of paint? a. (60%, 30%), b. (20%, 20%), c. (40%, 12%), d. (15%, 20%)

_Answer_ - a. (60, 30)

```python
import matplotlib.pyplot as plt
# help(pyplot)

U = (10, 10)
V = (60, 10)
W = (10, 30)

a = (60, 30)
b = (20, 20)
c = (40, 12)
d = (15,20)

points1 = [U, V, W]
points2 = [a, b, c, d]

x1, y1 = zip(*points1)
x2, y2 = zip(*points2)

plt.scatter(x1, y1, color='blue', marker='o', label='points')
plt.scatter(x2, y2, color='red', marker='o', label='points')
plt.show()



```

**Convex Schemes/Combinations**

- Convex Combination: points lie on the same segment
- Affine Combination: every point obtained by an affine combination lies on a line determined by the 2 points
- Hence it's implied that the collection of points obtained by Convex Combinations is a subset of that obtained by Affine Combinations
- Convex Independent: three points on a triangle ... a Convex Hull
- Convex Hull: the range bounded by the rubber band together with a number of nails.
